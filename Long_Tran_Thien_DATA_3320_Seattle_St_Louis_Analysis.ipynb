creating plots to answer the the question listed

```python
monthly_plot = sns.catplot(kind='bar', data=df, x='month', y='precipitation', hue='city', errorbar = None)
monthly_plot.set_xticklabels(["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
```

on average, we can see as the months changes St. Louis during the summer reachs its peak in avg. precipitation and drops during the winter, and the opposite happens for Seattle where Seattle has more precipitation during the Winter and drops in the Summer. Another detail to note is Seattle and St. Louis have differetst lowest drop points where Seattle drops down to 0.01 precipitation during the Summer while St. Louis only drops down to above 0.05 precipitation.

```python
sns.barplot(data = df, x = 'year', y = 'precipitation', hue = 'city', errorbar = None)
```

using avg. precipitation each year plot, we can see that St. Louis has more precipitation 4 out of 5 years vs Seattle
